```python
import cv2
from matplotlib import pyplot as plt
```

```python
fz = 16
# create figure
fig = plt.figure(figsize=(60, 50), dpi = 600)  # 60,50
  
# setting values to rows and column variables
rows =  2
columns = 1
  
Image1 = cv2.imread('Fig1a.png')
Image2 = cv2.imread('Fig1b.png')

fig.add_subplot(rows, columns, 1)
plt.imshow(Image1)
plt.axis('off')
#plt.title("Figure 1a", fontsize=fz)

fig.add_subplot(rows, columns, 2)
plt.imshow(Image2)
plt.axis('off')
#plt.title("Figure 1b", fontsize=fz)


fig.savefig("Fig1.pdf", dpi=600)
```
